```python
import cv2
import os
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

# Load dữ liệu và tính đặc trưng ảnh

```python
def TinhHist(pathfilename):
    img = cv2.imread(pathfilename,0)
    hist = cv2.calcHist([img],[0],None,[256],[0,256])
    size = img.shape[0]*img.shape[1]
    hist = hist / size
    return hist

def ReadData(path_to_file):
    features = []
    labels = []
    for file_name in sorted(os.listdir(path_to_file)):
        if file_name[:3] == 'cat':
            labels.append(0)
        else:
            labels.append(1)
        

        hist = TinhHist(os.path.join(path_to_file, file_name))
        features.append(hist)

    return features, labels
```

```python
DATA_DIR = '/content/drive/MyDrive/CS231.M22.KHCL/Baitap/Dataset/dogscats_small'
TRAIN_DIR = os.path.join(DATA_DIR, 'train')
TEST_DIR = os.path.join(DATA_DIR, 'test')
```

```python
train_features, train_labels = ReadData(TRAIN_DIR)
test_features, test_labels = ReadData(TEST_DIR)
```

# KNeighborsClassifier

```python
train_features = np.array(train_features)
train_features = np.squeeze(train_features)
train_labels = np.array(train_labels)

test_features = np.array(test_features)
test_features = np.squeeze(test_features)
test_labels = np.array(test_labels)
for i in range(1,6):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(train_features, train_labels)


    acc = model.score(test_features, test_labels)
    print(f'{i} neighbors => accuracy {acc}')
```
